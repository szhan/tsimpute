```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
```

```python
base_dir = "/Users/szhan/Projects/tsimpute/viz/"
out_png_file = base_dir + "sisu_venn.png"
dpi = 1_000
```

```python
sisu_42 = 8_554
sisu_3 = 3_775

sisu_3_unique = 1_209
overlap = sisu_3 - sisu_3_unique
sisu_42_unique = sisu_42 - overlap

venn2(
    subsets = (sisu_42_unique, sisu_3_unique, overlap),
    set_labels = (
        f"SiSu v4.2\nN = {sisu_42}",
        f"SiSu v3.0\nN = {sisu_3}",
    )
);

plt.savefig(out_png_file, dpi=dpi);
```
