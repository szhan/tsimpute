```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
base_dir = "/Users/szhan/Projects/tsimpute/viz/"
file_prefix = "imputation.results.leaves_kept"
in_csv_file = base_dir + file_prefix + ".csv"
out_png_file = base_dir + file_prefix + ".png"
```

```python
results = pd.read_csv(in_csv_file)
grid_data = results.pivot("mmr_ancestors", "mmr_samples", "perc_gt_90_iqs")
```

```python
plt.contourf(grid_data.columns, grid_data.index, grid_data, 20, cmap='viridis')
plt.xlabel("log10(mismatch ratio - sample haplotypes)")
plt.ylabel("log10(mismatch ratio - ancestor haplotypes)")
plt.xticks(ticks=np.arange(-7, 5))
plt.yticks(ticks=np.arange(-7, 5))
plt.colorbar(label="Percent of sites with IQS >= 0.9", location="top", orientation="horizontal")
plt.savefig(out_png_file, facecolor="white", dpi=1000)
```
