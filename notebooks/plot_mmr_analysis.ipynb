```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
base_dir = "/Users/szhan/Projects/tsimpute/viz/"
file_prefix = "mmr.imputation.results.leaves_kept"

in_csv_file = base_dir + file_prefix + ".csv"
out_png_file = base_dir + file_prefix + ".png"

dpi = 1_000
```

```python
results = pd.read_csv(in_csv_file)
grid_data = results.pivot("mmr_ancestors", "mmr_samples", "perc_gt_90_iqs")
```

```python
fig, ax = plt.subplots(figsize=(10, 10))

plt.contourf(grid_data.columns, grid_data.index, grid_data, 20, cmap='viridis')
plt.xticks(ticks=np.arange(-7, 5))
plt.yticks(ticks=np.arange(-7, 5))
cbar = plt.colorbar(location="top", orientation="horizontal")

ax.set_ylabel("log10(mismatch ratio - ancestor haplotypes)", size=20)
ax.set_xlabel("log10(mismatch ratio - sample haplotypes)", size=20)
ax.tick_params(axis='both', which='major', labelsize=20)
cbar.ax.set_title("% sites with IQS >= 0.9", size=20)
cbar.ax.tick_params(size=20, labelsize=20)

plt.savefig(out_png_file, facecolor="white", dpi=dpi)
```
